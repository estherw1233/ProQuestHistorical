```
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from pprint import pprint

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

```

```
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
t1 = "It hick inaugurates the new and welcome habit of selling bread by weight Lot every baker followed the people demand the the City Fathers ordain and the Police enforce it We have been too long fed on light and little loaves do"
t2 = "Your Apple Watch can also personalize your shoes. Instantly change the color of your lights or switch up your fit modes, all from your wrist."
pprint(preprocess(t2)) # list
```

```
sent  = preprocess(t2)
sent 

product =[]
for word in sent:
    if "NNP" in word:
        product.append(word[0])

print(", ".join(product))
# noun phrase: NP
# optional determiner: DT
# adjectives: JJ
# noun: NN.
```
